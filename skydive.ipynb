```python
import dataclasses

@dataclasses.dataclass

class Vec3:
    x: float
    y: float
    z: float

    def __add__(self, other):
        return Vec3(self.x + other.x, self.y + other.y, self.z + other.z)
    
    def __mul__(self, otherfloat):
        return Vec3(self.x * otherfloat, self.y * otherfloat, self.z * otherfloat)

```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temps = []
positions_x = []
positions_y = []
positions_z = []
vitesses_x = []
vitesses_y = []
vitesses_z = []
accelerations_x = []
accelerations_y = []
accelerations_z = []
```

```python
p0 = Vec3(0,0,0)
v0 = Vec3(100,0,100)
a = Vec3(0,0,-10)

t = 0
dt = 0.1

p = p0
v = v0

while t < 100:
    p = p + v * dt
    v = v + a * dt

    t = t + dt
    #print("temps = ",t)
    #print("position = ",p)
    #print("vitesse = ",v)
    print(f"{t} {p.x} {p.y} {p.z} {v.x} {v.y} {v.z}")

    
    # Sauvegarde des données pour affichage
    temps.append(t)
    positions_x.append(p.x)
    positions_y.append(p.y)
    positions_z.append(p.z)
    vitesses_x.append(v.x)
    vitesses_y.append(v.y)
    vitesses_z.append(v.z)
    accelerations_x.append(a.x)
    accelerations_y.append(a.y)
    accelerations_z.append(a.z)

```

```python
# Graphique 1 : Position en fonction du temps
plt.figure(1)
plt.plot(temps, positions_x, color='red', label='Position X')
plt.plot(temps, positions_y, color='blue', label='Position Y')
plt.plot(temps, positions_z, color='green', label='Position Z')


# Graphique 2 : Vitesse en fonction du temps
plt.figure(2)
plt.plot(temps, vitesses_x, label='Vitesse X')
plt.plot(temps, vitesses_y, label='Vitesse Y')
plt.plot(temps, vitesses_z, label='Vitesse Z')


# Graphique 3 : Accélération en fonction du temps
plt.figure(3)
plt.plot(temps, accelerations_x, label='Accélération X')
plt.plot(temps, accelerations_y, label='Accélération Y')
plt.plot(temps, accelerations_z, label='Accélération Z')


# Afficher les graphiques
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition de la classe Vec3
class Vec3:
    def __init__(self, x=0, y=0, z=0):
        self.x = x
        self.y = y
        self.z = z
    
    def __add__(self, other):
        return Vec3(self.x + other.x, self.y + other.y, self.z + other.z)
    
    def __sub__(self, other):
        return Vec3(self.x - other.x, self.y - other.y, self.z - other.z)
    
    def __mul__(self, scalar):
        return Vec3(self.x * scalar, self.y * scalar, self.z * scalar)
    
    def norm(self):
        return np.sqrt(self.x**2 + self.y**2 + self.z**2)
    
    def normalize(self):
        n = self.norm()
        if n == 0:
            return Vec3(0, 0, 0)
        return Vec3(self.x / n, self.y / n, self.z / n)
    
    def __repr__(self):
        return f"Vec3({self.x}, {self.y}, {self.z})"

# Données initiales
P0 = Vec3(0, 0, 4000)  # Position initiale en 3D (x, y, z)
V0 = Vec3(200 * 1000 / 3600, 0, 0)  # Vitesse initiale en 3D (200 km/h en m/s sur x)
m = 80  # Masse en kg
g = Vec3(0, 0, -10)  # Gravité agissant sur l'axe Z

# Conditions initiales
position = P0
velocity = V0
time_step = 0.1  # Pas de temps en secondes
max_time = 200  # Durée maximale de la simulation en secondes
coef = 0.32  # Coefficient de traînée
time = 0

# Pour stocker les données pour affichage
time_data = []
position_data = []
velocity_data_x = []
velocity_data_y = []
velocity_data_z = []
horizontal_distance_data = []

# Simulation
while position.z > 0:  # tant que l'altitude est positive
    # Force de gravité
    gravity_force = g * m
    
    # Force de traînée
    drag_force = velocity.normalize() * (coef * velocity.norm() ** 2)
    
    # Force nette
    net_force = gravity_force - drag_force
    
    # Accélération
    acceleration = net_force * (1 / m)
    
    # Mise à jour de la vitesse et de la position
    velocity = velocity + (acceleration * time_step)
    position = position + (velocity * time_step)
    
    # Calcul de la distance horizontale (projection sur le plan X-Y)
    horizontal_distance = np.sqrt(position.x**2 + position.y**2)
    
    # Stocker les données
    time_data.append(time)
    position_data.append([position.x, position.y, position.z])
    velocity_data_x.append(velocity.x)
    velocity_data_y.append(velocity.y)
    velocity_data_z.append(velocity.z)
    horizontal_distance_data.append(horizontal_distance)
    
    # Incrémenter le temps
    time += time_step

# Convertir les données pour affichage
position_data = np.array(position_data)
time_data = np.array(time_data)
velocity_data_x = np.array(velocity_data_x)
velocity_data_y = np.array(velocity_data_y)
velocity_data_z = np.array(velocity_data_z)
horizontal_distance_data = np.array(horizontal_distance_data)

# Affichage de la trajectoire en 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(position_data[:, 0], position_data[:, 1], position_data[:, 2], label='Trajectoire 3D')
ax.set_xlabel('Position X (m)')
ax.set_ylabel('Position Y (m)')
ax.set_zlabel('Altitude Z (m)')
ax.set_title('Simulation de la chute libre en 3D avec Vec3')
ax.grid(True)
plt.show()

# Affichage de la vitesse au cours du temps (vitesse globale)
plt.figure(figsize=(10, 6))
plt.plot(time_data, np.sqrt(velocity_data_x**2 + velocity_data_y**2 + velocity_data_z**2), label="Vitesse (m/s)", color="orange")
plt.xlabel("Temps (s)")
plt.ylabel("Vitesse (m/s)")
plt.title("Évolution de la vitesse du parachutiste")
plt.grid(True)
plt.show()

# Affichage des vitesses en fonction du temps pour chaque axe
plt.figure(figsize=(10, 6))
plt.plot(time_data, velocity_data_x, label="Vitesse sur X (m/s)", color='blue')
plt.plot(time_data, velocity_data_y, label="Vitesse sur Y (m/s)", color='green')
plt.plot(time_data, velocity_data_z, label="Vitesse sur Z (m/s)", color='red')
plt.xlabel("Temps (s)")
plt.ylabel("Vitesse (m/s)")
plt.title("Vitesse selon les axes X, Y, Z")
plt.legend()
plt.grid(True)
plt.show()

# Affichage de la hauteur en fonction de la distance horizontale
plt.figure(figsize=(10, 6))
plt.plot(horizontal_distance_data, position_data[:, 2], label="Hauteur (m)")
plt.xlabel("Distance horizontale (m)")
plt.ylabel("Hauteur (m)")
plt.title("Hauteur en fonction de la distance horizontale")
plt.grid(True)
plt.show()
```
